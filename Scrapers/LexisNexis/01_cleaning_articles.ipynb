```python
import pandas as pd
import re
import ast
import unicodedata
from tqdm import tqdm
from docx import Document
import locale
locale.setlocale(locale.LC_ALL, "nl_NL")
```

```python
with open('paper_selectors.txt', 'r') as file:
    papers = ast.literal_eval(file.read())

with open('party_selectors.txt', 'r') as file:
    parties = ast.literal_eval(file.read())

with open('months.txt', 'r') as file:
    months = ast.literal_eval(file.read())
```

```python
dates = pd.date_range('2023-08-21', '2023-10-08', freq='D').strftime('%Y-%m-%d')
```

Loading articles into lists

```python
articles = []

dates = pd.date_range('2023-10-02', '2023-10-03', freq='D').strftime('%Y-%m-%d')

for date in tqdm(dates):
    for n in range(8):
        try:
            docs = ""

            doc = Document(f"../../data/raw_dataset_new/{date} {n}.DOCX")
            for para in doc.paragraphs:
                docs = docs + para.text + " "

            # Split the text into individual articles based on "End of Document" separator, remove last empty article
            articles_otd = re.split(r'End of Document', docs)
            articles_otd = [article.split("\n") for article in articles_otd]
            articles_otd.pop(-1)
            articles_otd = [[article[1], article[3]] for article in articles_otd]
            articles_otd = [[articles[0], unicodedata.normalize('NFKD', articles[1])] for articles in articles_otd]
            print(articles_otd)
            articles.append(articles_otd)
        except:
            #TODO error catching
            print(f"{date} {n} coombination not found", end="/r")
```

Finding paper, title and date:

```python
text = articles[0][0][0]
print(text)
for paper in papers:
    for month in months:
        search = re.search(rf'\s(.*?)\s({paper})\s(\d+\s{month}\s2023)', text)
        if search is not None:
            print(search.groups())
```

Finding subject:

```python
text = articles[0][0][1]
print(text)
search = re.search(r'Section:\s(.*?);', text)
try:
    print(search.groups())
except:
    print(None)
```

Finding author:

```python
text = articles[0][0][1]
print(text)
search = re.search(r'Byline:\s(.*?)\s(Body|Highlight:)', text)
print(search.group(1))
```

Finding article body:

```python
text = articles[0][23][1]
print(text)
try:
    search = re.search(r'Body\s{3,}?(.*)\s+?(PDF-bestand|Bekijk\dhier|Bekijk\sde\soorspronkelijke\spagina|Link\snaar\sPDF)', text)
    search.groups() # Check if search fails
except:
    search = re.search(r'Body\s{3,}?(.*)', text)

print(search.groups())
```
